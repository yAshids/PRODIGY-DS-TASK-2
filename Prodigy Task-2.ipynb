```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
data
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = (
    data.dropna(subset=["Embarked"])
    .assign(Cabin=lambda x: x["Cabin"].fillna("Unknown"))
    .assign(Age=lambda x: x["Age"].fillna(x["Age"].mean()))
)
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 3))
sns.histplot(x="Age", data=data, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming `data` is your DataFrame
# Ensure the 'Survived' column is a category type for proper handling
data['Survived'] = data['Survived'].astype(str)

# Plot the data
sns.catplot(data=data, x="Sex", hue="Survived", kind="count", height=3, aspect=2)
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

# Manually setting the legend to avoid issues with automatic legend generation
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 3))
scatter_plot = sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived")
scatter_plot.set_title("Scatter Plot of Age vs Fare")
scatter_plot.set_xlabel("Age")
scatter_plot.set_ylabel("Fare")
scatter_plot.legend(title="Survived")
plt.show()
```
